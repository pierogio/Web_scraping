```python
# Se realiza la solicitud GET a la página web de Quotes to Scrape
# ==============================================================
my_request = requests.get('http://quotes.toscrape.com')

#Visualizamos status code e headers de la respuesta del servidor
print(my_request.status_code)
print(my_request.headers)
```

```python
# Visualizamos las cuotes de la primera pagina
# ================================================
doc = BeautifulSoup(my_request.text, "html.parser")

quotes = doc.select(".text")

for i in quotes:
    print ("Quote: ", i.text, "\n")
```

```python
# Visualizamos la primera quote
# ==============================
print (quotes[0].text)
```

```python
# Visualizamos los autores de la primera pagina
# =============================================
authors = doc.select(".author")

for i in authors:
    print ("Author:", i.text, "\n")
```

```python
# Visualizamos quote, author y tag de la pagina 1
# =============================================
final = doc.find_all("div", class_="quote")

url_base = "http://quotes.toscrape.com"

for i in final:
    quote = i.find("span", class_="text").text.strip()
    author = i.find("small", class_="author").text.strip()
    #tag = i.find("a", class_="tag").text.strip()
    tags = i.find_all("a", class_="tag")
    tag_list = [tag.text.strip() for tag in tags]
    tag = ', '.join(tag_list) 

    print("Quote:", quote)
    print("Author:", author)
    print("Tag:", tag)
    print()
```

```python
# Profundizamos y visualizamos quote, author y tag de todas las paginas de la web
# =================================================================================

class PostScraping():
    def __init__(self, quote, author, tag):
        self.quote = quote
        self.author = author
        self.tag = tag
    
class PostExtractor():    

    def extrae_info(self):
        
        url_base = "http://quotes.toscrape.com"
        
        post = []
        
        page_number = 1
        
        while url_base!="":

            my_doc = requests.get(url_base)
            
            doc_final = BeautifulSoup(my_doc.text, "html.parser")

            
            print(f"page : {page_number}")
            print("----------------------")
            
            for i in doc_final.select(".quote"):
                quote = i.select_one(".text").text
                author = i.select_one(".author").text
                tag = [tag.text.strip() for tag in i.select(".tags a")]


                scraper=PostScraping(quote, author, tag)
                post.append(scraper)
                
                print("Quote:", quote)
                print("Author:", author)
                print("Tags:", ", ".join(tag))
                print()

            next_page = doc_final.select_one(".next a")
            if next_page:
                rutas = urljoin(url_base, next_page.attrs["href"])
            else:
                rutas = ""
                
            url_base = rutas
            
            
            page_number += 1

            time.sleep(2)

        return post    
        
        
un_post=PostExtractor()

lista_post=un_post.extrae_info()
```

```python
# Creamos un df con los datos extraidos
# ====================================

data = {
    'Quote': [item.quote for item in lista_post],
    'Author': [item.author for item in lista_post],
    'Tags': [', '.join(item.tag) for item in lista_post]
}

df_quotes = pd.DataFrame(data)

df_quotes
```

```python
# Cargamos los drivers de selenium para chrome y hacemos scraping the precio y total reviews de monitores portatiles en Amazon
# ===========================================================================================================================

options = Options()
options.add_argument("--headless")
browser = webdriver.Chrome(options=options)
url = 'https://www.amazon.es/s?k=monitor+port%C3%A1til&crid=2VUJFUHL6QAPZ&sprefix=%2Caps%2C107&ref=nb_sb_ss_recent_1_0_recent'
browser.get(url)
browser.maximize_window()

result_items = browser.find_elements(
    by=By.CSS_SELECTOR,
    value='.s-result-item'
)

data = []

for product in result_items:
    try:
        product_title = product.find_element(
            by=By.CSS_SELECTOR,
            value='.s-title-instructions-style'
        )
    except NoSuchElementException:
        continue

    try:
        product_price_element = product.find_element(
            by=By.CSS_SELECTOR,
            value='.a-price-whole'
        )
    except NoSuchElementException:
        continue

    product_price = product_price_element.text

    try:
        product_review_element = product.find_element(
            by=By.CSS_SELECTOR,
            value='.a-size-base.s-underline-text'
        )
    except NoSuchElementException:
        continue

    product_review = product_review_element.text
    
    data.append({
        'Title': product_title.text,
        'Price': product_price,
        'Review': product_review
    })

browser.quit()

df_amazon = pd.DataFrame(data)

df_amazon
```

```python
# Creamos un spider basico para extraer precios y datos de monitores portatiles de Amazon con Scrapy
# Visualizamos solamente los monitores que tengan algun tipo de descuento
# ===================================================================================================

results = []

class AmazonSpider(scrapy.Spider):
    name = "amazon"
    start_urls = ["https://www.amazon.es/s?k=monitor+port%C3%A1til&crid=2VUJFUHL6QAPZ&sprefix=%2Caps%2C107&ref=nb_sb_ss_recent_1_0_recent"]
    custom_settings = {
            "USER_AGENT": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36",
            "CONCURRENT_REQUESTS": 1,
            "DOWNLOAD_DELAY": 5
        }
    

    def parse(self, response):
        for product in response.css(".s-result-item"):
            data = {
                "title": product.css(".a-link-normal .a-text-normal::text").get(),
                "price": product.css(".a-price .a-price-whole::text").get(),
                "rating": product.css(".a-icon .a-icon-alt::text").get(),
                "reviews": product.css(".a-link-normal .a-size-base::text").get(),
                "coupon" : product.css(".s-coupon-unclipped .a-size-base::text").get()
                }
            results.append(data)

process = CrawlerProcess(get_project_settings())
process.crawl(AmazonSpider)
process.start()

df_amazon_final = pd.DataFrame(results)
df_amazon_final[df_amazon_final["coupon"].notnull()]
```
